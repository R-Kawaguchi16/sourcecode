```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense
import sentencepiece as spm
import numpy as np
import sys
from dictionary import p_dic, reverse_p_dic
from inSplit import input_list, input_max_pieces
from outSplit import output_list, output_max_pieces
import pandas as pd

batch_size = 64 # Batch size for training.
epochs = 200  # Number of epochs to train for.
latent_dim = 256  # Latent dimensionality of the encoding space.
num_samples = 20000  # Number of samples to train on.
# Path to the data txt file on disk.
#data_path = 'input.txt'
 
# Vectorize the data.
lines = len(input_list)
pieces = len(p_dic)
input_vec = np.zeros((lines,input_max_pieces,pieces),dtype = 'int8')
output_vec = np.zeros((lines,output_max_pieces,pieces),dtype = 'int8')
target_vec = np.zeros((lines,output_max_pieces,pieces),dtype = 'int8')

for i,(In,Out) in enumerate(zip(input_list,output_list)):
    for t,piece in enumerate(In):
        input_vec[i,t,p_dic[piece]] = 1
    for t,piece in enumerate(Out):
        output_vec[i,t,p_dic[piece]] = 1
        if t > 0:
            target_vec[i,t-1,p_dic[piece]] = 1
#input_texts = []
#target_texts = []
#input_characters = set()
#target_characters = set()
#with open(data_path, 'r', encoding='utf-8') as f:
#    lines = f.read().split('\n')
#for line in lines[: min(num_samples, len(lines) - 1)]:
#    input_text, target_text = line.split('\t')
    # We use "tab" as the "start sequence" character
    # for the targets, and "\n" as "end sequence" character.
#    target_text = '\t' + target_text + '\n'
#    input_texts.append(input_text)
#    target_texts.append(target_text)
#    for char in input_text:
#        if char not in input_characters:
#            input_characters.add(char)
#    for char in target_text:
#        if char not in target_characters:
#            target_characters.add(char)

#input_characters = sorted(list(input_characters))
#target_characters = sorted(list(target_characters))
#num_encoder_tokens = len(input_characters)
#num_decoder_tokens = len(target_characters)
#max_encoder_seq_length = max([len(txt) for txt in input_texts])
#max_decoder_seq_length = max([len(txt) for txt in target_texts])

#print('input_texts\n',input_texts)               #list(str)
#print('\ntarget_texts\n',target_texts)           #list(\t+str+\n)
#print('\ninput_characters\n',input_characters)   #input_textsに含まれる全ての文字のlist
#print('\ntarget_characters\n',target_characters) #target_textsに含まれる全ての文字のlist

print('Number of samples:', lines)
print('Number of unique pieces:', pieces)
#print('Number of unique output tokens:', num_decoder_tokens)
print('Max pieces length for inputs:', input_max_pieces)
print('Max pieces length for outputs:', output_max_pieces)
 
#input_token_index = dict(
#    [(char, i) for i, char in enumerate(input_characters)])
#target_token_index = dict(
#    [(char, i) for i, char in enumerate(target_characters)])  #辞書作成
 
#encoder_input_data = np.zeros(
#    (len(input_texts), max_encoder_seq_length, num_encoder_tokens),dtype = 'float32')
#decoder_input_data = np.zeros(
#    (len(input_texts), max_decoder_seq_length, num_decoder_tokens),dtype = 'float32')
#decoder_target_data = np.zeros(
#    (len(input_texts), max_decoder_seq_length, num_decoder_tokens),dtype = 'float32')
#要素が全て0の3次元配列
#encoder_input_data:20000*37*126
#decoder_input_data:20000*50*126
#decoder_target_data:20000*50*126
 
#for i, (input_text, target_text) in enumerate(zip(input_texts, target_texts)): #12/14の研究ノート
#    for t, char in enumerate(input_text): #1文字ずつ
#        encoder_input_data[i, t, input_token_index[char]] = 1. #i番目のinputのt文字目がどの文字か
#    for t, char in enumerate(target_text):
        # decoder_target_data is ahead of decoder_input_data by one timestep
#        decoder_input_data[i, t, target_token_index[char]] = 1.
#        if t > 0:
            # decoder_target_data will be ahead by one timestep
            # and will not include the start character.
#            decoder_target_data[i, t - 1, target_token_index[char]] = 1.
            #decoder_targetはdecoder_inputより一つ進んでいる

encoder_inputs = Input(shape=(None, pieces))
encoder = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
# We discard `encoder_outputs` and only keep the states.
encoder_states = [state_h, state_c]
 
# Set up the decoder, using `encoder_states` as initial state.
decoder_inputs = Input(shape=(None, pieces))
# We set up our decoder to return full output sequences,
# and to return internal states as well. We don't use the
# return states in the training model, but we will use them in inference.
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs,
                                     initial_state=encoder_states)
decoder_dense = Dense(pieces, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)
 
# Define the model that will turn
# `encoder_input_data` &amp; `decoder_input_data` into `decoder_target_data`
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
 
# Run training
model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
 
model.summary()  # スクリーンにモデル表示
 
model.fit([input_vec, output_vec], target_vec,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.2)
# Save model
model.save('s2s.h5')
 
# Next: inference mode (sampling).
# Here's the drill:
# 1) encode input and retrieve initial decoder state
# 2) run one step of decoder with this initial state
# and a "start of sequence" token as target.
# Output will be the next target token
# 3) Repeat with the current target token and current states
 
# Define sampling models
encoder_model = Model(encoder_inputs, encoder_states)
 
decoder_state_input_h = Input(shape=(latent_dim,))
decoder_state_input_c = Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_inputs, initial_state=decoder_states_inputs)
decoder_states = [state_h, state_c]
decoder_outputs = decoder_dense(decoder_outputs)
decoder_model = Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs] + decoder_states)
 
# Reverse-lookup token index to decode sequences back to
# something readable.
#reverse_input_char_index = dict(
#    (i, char) for char, i in input_token_index.items())
#reverse_target_char_index = dict(
#    (i, char) for char, i in target_token_index.items())
 
def decode_sequence(input_seq):
    # Encode the input as state vectors.
    states_value = encoder_model.predict(input_seq)
 
    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1, 1, pieces))
    # Populate the first character of target sequence with the start character.
    target_seq[0, 0, p_dic['\t']] = 1
 
    # Sampling loop for a batch of sequences
    # (to simplify, here we assume a batch of size 1).
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value)
 
        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_piece = reverse_p_dic[sampled_token_index]
        decoded_sentence += sampled_piece
 
        # Exit condition: either hit max length
        # or find stop character.
        if (sampled_piece == '\n' or
           len(decoded_sentence) > output_max_pieces):
            stop_condition = True
 
        # Update the target sequence (of length 1).
        target_seq = np.zeros((1, 1, pieces))
        target_seq[0, 0, sampled_token_index] = 1
 
        # Update states
        states_value = [h, c]
 
    return decoded_sentence

test = open('test.txt','r',encoding='utf8')
sp = spm.SentencePieceProcessor()
sp.Load("split.model")
all_in = []
all_tar = []
all_out = []
#f = open("seq2seq.txt","w",encoding="utf-8") 
for line in test:
    # Take one sequence (part of the training set)
    # for trying out decoding.
    In,Out = line.split('\t')
    all_in.append(In)
    all_tar.append(Out)
    input_seq = np.zeros((1,input_max_pieces,pieces),dtype = 'int8')
    sequence = sp.EncodeAsPieces(In)
    if sequence[0] == '▁':
        del sequence[0]
    else:
        sequence[0] = sequence[0][1:]
    for i,piece in enumerate(sequence):
        input_seq[0,i,p_dic[piece]] = 1
    #print('input_seq.shape:',input_seq.shape)
    #print('input_seq:',input_seq)
    decoded_sentence = decode_sequence(input_seq)
    all_out.append(decoded_sentence)

df = pd.DataFrame({'input(入力)':all_in, 'target(正解)':all_tar, 'output(出力)':all_out})
df.to_csv('result_e200_b64_d20.tsv',index=False,sep='\t')

    #f.write('Input sentence:'+input_texts[seq_index]+'\nDecoded sentence:'+decoded_sentence+'\n')
#f.close()
```

```python
import os
import sys
import pandas as pd

def extract_text(filename):
    with open(filename) as f:
        text = f.readlines()
        text = [sentence.strip() for sentence in text]
        text = list(filter(lambda line: line != '', text))
        return text

all_input = []
all_target = []

f = extract_text('input.txt')

for line in f:
    line = line.split('\t')
    input_text = line[0]
    target_text = line[1]
    all_input.append(input_text)
    all_target.append(target_text)
print(len(all_input))
print(len(all_target))
```

```python
df = pd.DataFrame({'input':all_input, 'target':all_target})
df = df.sample(frac=1,random_state=2434).reset_index(drop=True)
```

```python
#入力・出力分割(inSplit.py,outSplit.py)、辞書作成(dictionary.py)
import sentencepiece as spm
import pprint

sp = spm.SentencePieceProcessor()
sp.Load("split.model")
line = open('train.txt','r',encoding='utf8').readlines()
line2 = open('test.txt','r',encoding='utf8').readlines()
input_list = []
output_list = []
input_max_pieces = 0
output_max_pieces = 0
p_set = set()
for s in line:
	Input,Output = s.split('\t')
	list_in = sp.EncodeAsPieces(Input)
	list_out = sp.EncodeAsPieces(Output)
	if list_in[0] == '▁':
		del list_in[0]
	else:
		list_in[0] = list_in[0][1:] 
	if list_out[0] == '▁':
		del list_out[0]
	else:
		list_out[0] = list_out[0][1:]
#	print(list_in)
#	print(list_out)
	list_out.insert(0,'\t')
	list_out.append('\n')
	input_list.append(list_in)
	output_list.append(list_out)
#for In,Out in zip(input_list,output_list):
	if len(list_in) > input_max_pieces:
		input_max_pieces = len(list_in)
	if len(list_out) > output_max_pieces:
		output_max_pieces = len(list_out)
	for p in list_in:
		if p not in p_set:
			p_set.add(p)
	for p in list_out:
		if p not in p_set:
			p_set.add(p)

for s in line2:
	fir,sec = s.split('\t')
	l1 = sp.EncodeAsPieces(fir)
	l2 = sp.EncodeAsPieces(sec)
	if l1[0] == '▁':
		del l1[0]
	else:
		l1[0] = l1[0][1:] 
	if l2[0] == '▁':
		del l2[0]
	else:
		l2[0] = l2[0][1:]
	if len(l1) > input_max_pieces:
		input_max_pieces = len(l1)
	if len(l2) > output_max_pieces:
		output_max_pieces = len(l2)
	for p in l1:
		if p not in p_set:
			p_set.add(p)
	for p in l2:
		if p not in p_set:
			p_set.add(p)

p_list = sorted(list(p_set))
p_dic = dict([(p,i) for i,p in enumerate(p_list)])
print(len(p_dic))
#lines = len(input_list)
pieces = len(p_dic)
reverse_p_dic = dict((i,char) for char,i in p_dic.items())

f1 = open('inSplit.py','w',encoding='utf8')
f2 = open('outSplit.py','w',encoding='utf8')
f3 = open('dictionary.py','w',encoding='utf8')
f1.write('input_list = ' + pprint.pformat(input_list) + '\n' + 'input_max_pieces = ' + pprint.pformat(input_max_pieces) + '\n')
f2.write('output_list = ' + pprint.pformat(output_list) + '\n' + 'output_max_pieces = ' + pprint.pformat(output_max_pieces) + '\n')
f3.write('p_dic = ' + pprint.pformat(p_dic) + '\n' + 'reverse_p_dic = ' + pprint.pformat(reverse_p_dic) + '\n' + 'pieces = ' + pprint.pformat(pieces) + '\n')
f1.close()
f2.close()
f3.close()

#input_vec = np.zeros((lines,input_max_pieces,pieces),dtype = 'float32')
#output_vec = np.zeros((lines,output_max_pieces,pieces),dtype = 'float32')

#for i,(In,Out) in enumerate(zip(input_list,output_list)):
#	for t,piece in enumerate(In):
#		input_vec[i,t,p_dic[piece]] = 1.
#	for t,piece in enumerate(Out):
#		output_vec[i,t,p_dic[piece]] = 1.


#o = input()
```
